```python
import numpy as np
import matplotlib.pyplot as plt
import toml
import os
import sys

module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

from sta_analysis.receptive_field_analysis import *
from sta_analysis.util_plotting import *
from sta_analysis.correlations import *
```

```python
data_dir = ''
gauss_spike_train = np.load(os.path.join(data_dir, 'spike_train_gauss_1.npy'))
gauss_stimulus = np.load(os.path.join(data_dir, 'stimulus_gauss_1.npy'))
gauss_filter = np.load(os.path.join(data_dir, f'filter_gauss_1.npy'))
gauss_stimulus = gauss_stimulus.transpose((1,2,0))

gabor_spike_train = np.load(os.path.join(data_dir, 'spike_train_gabor_1.npy'))
gabor_stimulus = np.load(os.path.join(data_dir, 'stimulus_gabor_1.npy'))
gabor_filter = np.load(os.path.join(data_dir, f'filter_gabor_1.npy'))
gabor_stimulus = gabor_stimulus.transpose((1,2,0))

ricker_spike_train = np.load(os.path.join(data_dir, 'spike_train_ricker_1.npy'))
ricker_stimulus = np.load(os.path.join(data_dir, 'stimulus_ricker_1.npy'))
ricker_filter = np.load(os.path.join(data_dir, f'filter_ricker_1.npy'))
ricker_stimulus = ricker_stimulus.transpose((1,2,0))

config_file = "config.toml"
with open(config_file, "r") as f:
    config = toml.load(f)

top_n = 3
```

```python
gauss = STA(gauss_stimulus, gauss_spike_train, gauss_filter)
gauss.calc_sta()
gauss.calc_stc_simplified()

gauss.plot_eigenvectors_all_lags(top_n = top_n)
gauss.plot_eigenvalue_spectrum(3)

plot_simple_sta_lags(gauss.sta, gauss.filter, gauss.start, gauss.end)
plot_mean_sta_with_filter(gauss.mean_sta, gauss.filter)

print(calc_pearson(gauss.mean_sta, gauss.filter))
print(calc_cosine_similarity(gauss.mean_sta, gauss.filter))
print(calc_mse(gauss.mean_sta, gauss.filter))
```

```python
ricker = STA(ricker_stimulus, ricker_spike_train, ricker_filter)
ricker.calc_sta()
ricker.calc_stc_simplified()
ricker.plot_eigenvectors_all_lags(top_n = top_n)

plot_simple_sta_lags(ricker.sta, ricker.filter, ricker.start, ricker.end)
plot_mean_sta_with_filter(ricker.mean_sta, ricker.filter)

print(calc_pearson(ricker.mean_sta, ricker.filter))
print(calc_cosine_similarity(ricker.mean_sta, ricker.filter))
print(calc_mse(ricker.mean_sta, ricker.filter))
```

```python
gabor = STA(gabor_stimulus, gabor_spike_train, gabor_filter)
gabor.calc_sta()
gabor.calc_stc_simplified()
gabor.plot_eigenvectors_all_lags(top_n = top_n)

plot_simple_sta_lags(gabor.sta, gabor.filter, gabor.start, gabor.end)
plot_mean_sta_with_filter(gabor.mean_sta, gabor.filter)

print(calc_pearson(gabor.mean_sta, gabor.filter))
print(calc_cosine_similarity(gabor.mean_sta, gabor.filter))
print(calc_mse(gabor.mean_sta, gabor.filter))
```
